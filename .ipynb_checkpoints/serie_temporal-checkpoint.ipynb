```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from datetime import datetime

# Registro dos converters p/ plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
# Carregando as bases de dados
df = pd.read_csv('AirPassengers.csv')
df.head()
```

```python
df.dtypes
```

```python
# lambda para converter e separar os valores
dateparse = lambda dates: datetime.strptime(dates, '%Y-%m')

#nova criação, passando a expressão para transformar a data
base = pd.read_csv('AirPassengers.csv', # o csv
                   parse_dates=['Month'], # coluna a ser transformada
                   index_col= 'Month', # index da coluna
                   date_parser=dateparse)
base.info()
```

```python
# vendo o index do df
base.index
```

```python
# criando a serie temporal

ts = base['#Passengers']
ts
```

```python
#vendo um período específico
ts['1949-01':'1949-05']
```

```python
#vendo 49 e 50
ts['1949':'1950']
```

```python
# vendo um ano
ts['1950']
```

```python
# vendo um ano
ts[:'1950']
```

```python
# Vendo ultimo index - "Maior" ou "mais novo"
ts.index.max()
```

```python
# Vendo primeiro index - "menor" ou "mais antigo"
ts.index.min()
```

```python
# Vendo toda serie temporal

plt.figure(figsize=(16,5))
plt.title("Série temporal 1949 até 1950")
plt.xlabel('Tempo')
plt.ylabel('Passagens')
plt.plot(ts);
```

```python
# Vendo toda serie temporal

plt.figure(figsize=(16,5))
plt.title("Série temporal 1949")
plt.xlabel('Tempo')
plt.ylabel('Passagens')
plt.plot(ts['1949'], color='cyan');
```

```python
# Vendo toda serie temporal

plt.figure(figsize=(16,5))
plt.title("Série temporal 1949")
plt.xlabel('Tempo')
plt.ylabel('Passagens')
plt.plot(ts['1950':'1955']);
```

```python
# Vendo por ano

# Vendo toda serie temporal

plt.figure(figsize=(16,5))
plt.title("Série temporal 1949")
plt.xlabel('Tempo')
plt.ylabel('Passagens')
plt.plot(ts.resample('A').sum());
```

```python
# agrupando por mes 
ts_mes = ts.groupby([lambda x : x.month]).sum()
plt.figure(figsize=(16,5))
plt.title("Série temporal 1949")
plt.xlabel('Tempo')
plt.ylabel('Passagens')
plt.plot(ts_mes);
```
