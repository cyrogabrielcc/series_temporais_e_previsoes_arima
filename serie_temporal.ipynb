```python
# lambda para converter e separar os valores
dateparse = lambda dates: datetime.strptime(dates, '%Y-%m')

#nova criação, passando a expressão para transformar a data
base = pd.read_csv('AirPassengers.csv', # o csv
                   parse_dates=['Month'], # coluna a ser transformada
                   index_col= 'Month', # index da coluna
                   date_parser=dateparse)
base.info()
```

```python
# Vendo por ano

# Vendo toda serie temporal

plt.figure(figsize=(16,5))
plt.title("Série temporal 1949")
plt.xlabel('Tempo')
plt.ylabel('Passagens')
plt.plot(ts.resample('A').sum());
```

```python
# agrupando por mes 
ts_mes = ts.groupby([lambda x : x.month]).sum()
plt.figure(figsize=(16,5))
plt.title("Série temporal 1949")
plt.xlabel('Tempo')
plt.ylabel('Passagens')
plt.plot(ts_mes);
```

### Decomposição

```python
# Importando a lib
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Decomposição da Série temporal - Criando uma variável para cada formato
decomposicao = seasonal_decompose(ts)
decomposicao
```

```python
# Tendência
tendencia = decomposicao.trend
tendencia
```

```python
# Sazonalidade
sazonal = decomposicao.seasonal
sazonal
```

```python
aleatorio = decomposicao.resid
aleatorio
```

```python
plt.figure(figsize=(7,10))

plt.subplot(4,1,1)
plt.plot(ts, label='Original')
plt.legend(loc='best')

#vendo apenas a tendencicia
plt.subplot(4,1,2)
plt.plot(tendencia, label='Tendência')
plt.legend(loc='best')

# Sazonalidade
plt.subplot(4,1,3)
plt.plot(tendencia, label = "Sazonalidade")
plt.legend(loc='best')

# Elemento aleatorio
plt.subplot(4,1,4)
plt.plot(aleatorio, label = "aleatorio")
plt.legend(loc='best')

plt.tight_layout()
```
